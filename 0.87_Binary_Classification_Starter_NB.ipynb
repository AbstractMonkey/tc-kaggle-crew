```python
df.drop(columns=['id'], inplace=True)
x=df.iloc[:,:-1]
y=df.loc[:,'Response']
```

```python
obj_col = x.select_dtypes(include=['object']).columns

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
for col in obj_col:
    x[col] = le.fit_transform(x[col])
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
x=scaler.transform(x)
```

```python
from sklearn.model_selection import RandomizedSearchCV,train_test_split
from xgboost import XGBClassifier
from catboost import CatBoostClassifier
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
xgb_best = XGBClassifier(
    use_label_encoder=False, 
    eval_metric='auc',
    subsample=1.0,
    n_estimators=400,
    max_depth=6,
    learning_rate=0.2,
    colsample_bytree=0.7
)
xgb_best.fit(x,y)
```

```python
catboost_best = CatBoostClassifier(
    silent=True,
    l2_leaf_reg=9,
    iterations=300,
    depth=6,
    learning_rate=0.1,
)
catboost_best.fit(x,y)
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.metrics import roc_auc_score

voting_clf = VotingClassifier(estimators=[
    ('xgb', xgb_best),
    ('catboost', catboost_best)
], voting='soft')

voting_clf.fit(X_train, y_train)

y_pred_proba = voting_clf.predict_proba(X_test)[:, 1]
roc_auc_catboost = roc_auc_score(y_test, y_pred_proba)
```

```python
test=pd.read_csv('/kaggle/input/playground-series-s4e7/test.csv')
temp=test['id']
test.drop(columns=['id'],inplace=True)

for col in obj_col:
    test[col] = le.fit_transform(test[col])
    
test=scaler.fit_transform(test)

y_pred=voting_clf.predict_proba(test)[:, 1]

z = pd.DataFrame({
    'id': temp,
    'Response': y_pred
})

z.to_csv('/kaggle/working/submission.csv', index=False)
```
